```python
import pandas as pd
import numpy as np
from sklearn.decomposition import TruncatedSVD
```

```python
ratings = pd.read_csv('~/Downloads/ratings.csv')
```

```python
books = pd.read_csv('~/Downloads/books.csv')
```

```python
ratings.head(1)
```

```python
books.head(1)
```

```python
columns = ['id','book_id', 'best_book_id', 'work_id', 'books_count', 'isbn', 'isbn13', 'authors', 'original_publication_year', 'original_title', 'title', 'language_code', 'average_rating', 'ratings_count', 'work_ratings_count', 'work_text_reviews_count', 'ratings_1', 'ratings_2', 'ratings_3', 'ratings_4', 'ratings_5', 'image_url', 'small_image_url']
```

```python
combined_books_data = pd.merge(ratings, books, on='book_id')
```

```python
combined_books_data.head(1)
```

```python
rating_utility_matrix = combined_books_data.pivot_table(values = 'rating', index = 'user_id', columns = 'title', fill_value = 0)
```

```python
rating_utility_matrix.head(2)
```

```python
rating_utility_matrix.shape
```

```python
X = rating_utility_matrix.T
```

```python
X.shape
```

```python
SVD = TruncatedSVD(n_components=30)
```

```python
SVD
```

```python
transposed_matrix = SVD.fit_transform(X)
```

```python
transposed_matrix.shape
```

```python
corr_matrix = np.corrcoef(transposed_matrix)
```

```python
books = rating_utility_matrix.columns
```

```python
books.shape
```

```python
books_list = list(books)
```

```python
book_1984 = books_list.index("1984")
```

```python
book_1984
```

```python
corr_1984 = corr_matrix[book_1984]
```

```python
list(books[(corr_1984<=1.0) & (corr_1984>=0.8)])
```
